```python
#creating a tags column by combining all relevant information
movies_final['tags']=movies_final['overview']+movies_final['genres']+movies_final['cast']+movies_final['crew']+movies_final['keywords']
```

```python
new_df=movies_final[['id_x', 'title', 'tags']]
```

```python
#convert list to string for tags
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
```

```python
#convert tags to lower case for simplicity

new_df['tags'] =new_df['tags'].apply(lambda x: x.lower())
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
  y=[]

  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y) #convert back to string after stemming
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer #basically counts the frequency of words in each piece of text

cv=CountVectorizer(max_features=4000, stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
cv.get_feature_names_out()
```

```python
#there are some words like 'action', 'actions' and similar sow e apply something called stemming where we replace all similar words with one word
!pip install nltk
```

```python
#we have to calculate the distance  between two vectors, lower distance is better
#we are using cosine distance
#euclidean distance fails in higher dimensioins

from sklearn.metrics.pairwise import cosine_similarity

similar=cosine_similarity(vectors) #this shoes the similarity between first movie with every other movies so obviouslt its tarts with one since first movie is absolutely similar with itself
```

```python
def recommend(movie):
  movie_index= new_df[new_df['title']==movie].index[0] #here we are trying to search for a movie given by user say 'avatar'
  distances=similar[movie_index] #IF THE MOVIE WE ARE SEARCHING IS AVATAR then we extract its similarity score with all other 4000 words
  movies_list= sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:7]
  #OKAY here we need top movies closest to avatar. Before sortinh though we have to assign an index to each score say 1. for first movie, 2. for second
  #otherwise we sort and the index gets lost, so then we sort based on scores while the numbers are still there
  #we sort based on second thing which is value (enumertae creates a key and value kind of this ensures we only arrange the vale\
  #then we extract the top 6 changed this was 5 in origina
  for i in movies_list:
    print(new_df.iloc[i[0]].title) #gives me movies index
```

```python
recommend('Batman Begins')
```

```python
recommend('Newlyweds')
```

```python
recommend('Avatar')
```
